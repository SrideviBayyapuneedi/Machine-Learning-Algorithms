# Support Vector Machine

This example does not use cross validation, we have to use cross validation to check how the model performs

Source:: http://scikit-learn.org/stable/modules/svm.html

### Steps to be followed:

1. Import library
2. Load the data, scale it if required
3. Object creation, specifying the model parameters like C and gamma
4. Fitting model
5. Prediction

### Dataset:
Iris dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
```

The different parameters that can be used for tuning an SVM

sklearn.svm.SVC(C=1.0,
                kernel='rbf',

```python
# import some data to play with
iris = datasets.load_iris()
#iris data set has features - sepal length, sepal width, petal length, petal width, target, target_names
```

```python
X = iris.data[:, :2] # we only take the first two features
y = iris.target
X.flags # make the data C-Contiguous to improve performance
```

```python
# we create an instance of SVM and fit out data. We do not scale our
# data since we want to plot the support vectors
C = 1.0 # SVM regularization parameter
svc = svm.SVC(kernel='linear', C=1,gamma=0).fit(X, y)
```

```python
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))
```

```python
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with linear kernel')
plt.show()
```

Now, let us see how rbf kernel works. It is used for mapping nonlinear relationships.
In this example we start by tuning the model's kernel coefficient.
*** As the value of gamma increases we can see that the model overfits to the data. We have to use cross validation techniques to avoid regularization error. ***

```python
#using rbf kernel with gamma (Kernel coefficient) = 0 
svc = svm.SVC(kernel='rbf', C=1,gamma=0).fit(X, y)
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with rbf kernel')
plt.show()
```

```python
#using rbf kernel with gamma (Kernel coefficient) = 10 
svc = svm.SVC(kernel='rbf', C=1,gamma=10).fit(X, y)
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with rbf kernel')
plt.show()
```
